```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the original dataset
new_df = pd.read_csv("Data/data.csv")

# Separate the features (X) and the target variable (y)
X = new_df.drop("Bankrupt?", axis=1).values
y = new_df["Bankrupt?"].values

# Create a DataFrame using the feature array
features_df = pd.DataFrame(X, columns=new_df.drop("Bankrupt?", axis=1).columns)

# Plot correlation matrix
correlation_matrix = features_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Load the preprocessed dataset
new_df = pd.read_csv("Data/sorted_correlation_with_values.csv")

# Separate the features (X) and the target variable (y)
X = new_df.drop("Bankrupt?", axis=1).values
y = new_df["Bankrupt?"].values

# Create a DataFrame using the feature array
features_df = pd.DataFrame(X, columns=new_df.drop("Bankrupt?", axis=1).columns)

# Calculate summary statistics
summary_stats = features_df.describe()
```

```python
# Plot pie chart for the target variable
target_counts = pd.Series(y).value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%')
plt.title('Target Variable (Bankrupt) Distribution')
plt.show()
```

```python
# Plot correlation matrix
correlation_matrix = features_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot box plots for each numerical feature
fig, axs = plt.subplots(nrows=len(features_df.columns), figsize=(8, 6 * len(features_df.columns)))
for i, col in enumerate(features_df.columns):
    axs[i].boxplot(features_df[col], vert=False)
    axs[i].set_title(col + ' - Box Plot')
    axs[i].set_xlabel('Value')

plt.tight_layout()
plt.show()
```

```python
# Plot histograms for each numerical feature
fig, axs = plt.subplots(nrows=len(features_df.columns), figsize=(8, 6 * len(features_df.columns)))
for i, col in enumerate(features_df.columns):
    axs[i].hist(features_df[col], bins=30, edgecolor='black')
    axs[i].set_title(col + ' - Histogram')
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
