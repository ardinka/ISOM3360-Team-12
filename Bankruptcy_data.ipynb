```python
drop_columns=[" Total income/Total expense"," Operating Profit Growth Rate"," Liability to Equity"," Net Income to Stockholder's Equity"," Gross Profit to Sales"," Cash Flow to Equity"," Equity to Long-term Liability"," Current Liability to Equity"," Cash Flow to Sales"," Retained Earnings to Total Assets"," Working Capital/Equity"," Operating Funds to Liability"," Inventory and accounts receivable/Net value"," Operating profit/Paid-in capital"," Net Value Growth Rate"," Total Asset Return Growth Rate Ratio"," Total Asset Growth Rate"," After-tax Net Profit Growth Rate"," Operating Profit Per Share (Yuan ¥)"," Cash Flow Per Share"," Realized Sales Gross Margin"," Non-industry income and expenditure/revenue"," ROA(C) before interest and depreciation before interest"," ROA(B) before interest and depreciation after tax", " Pre-tax net Interest Rate", " Net Value Per Share (C)"," Liability-Assets Flag"," Net Income Flag"," Net Value Per Share (B)"," Continuous interest rate (after tax)"," Per Share Net profit before tax (Yuan ¥)", " Net profit before tax/Paid-in capital"," Regular Net Profit Growth Rate"," Continuous Net Profit Growth Rate"," Revenue per person"," Operating profit per person"," Allocation rate per person"," Average Collection Days"," Inventory/Working Capital"," Inventory/Current Liability"," Long-term Liability to Current Assets"]
bankruptcy_data_df=bankruptcy_data_df.drop(columns=drop_columns)
```

```python
null_values = bankruptcy_data_df.isnull().sum()
null_values
```

```python
minmax_col=[" Operating Expense Rate"," Research and development expense rate"," Inventory Turnover Rate (times)"," Fixed Assets Turnover Frequency"," Quick Asset Turnover Rate"," Current Asset Turnover Rate"," Cash Turnover Rate"]
for col in minmax_col:
  minmax_scaler = preprocessing.MinMaxScaler().fit(bankruptcy_data_df[[col]])
  bankruptcy_data_df[col] = minmax_scaler.transform(bankruptcy_data_df[[col]])
```
